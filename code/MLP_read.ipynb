```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
x=pd.read_csv('iris.csv')
print(x.info())
print(x.describe())
print(x.head(3))
print(x.tail(3))
print(type(x))
data=x.dropna()
print(data)
print(x.duplicated().any())
print(x.drop_duplicates())
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
y=x.iloc[:,:4].values
z=x.iloc[:,4].values
print("data",y)
print("targt",z)
le=LabelEncoder()
zn=le.fit_transform(z)
print(zn)
ytrain,ytest,ztrain,ztest=train_test_split(y,zn,test_size=.25,random_state=1)
print(ztest)
mlp=MLPClassifier(hidden_layer_sizes=(30),activation='relu')
mlp.fit(ytrain,ztrain)
zpredict=mlp.predict(ytest)
print("testcls:-----",ztest)
print("prdcls:------",zpredict)
print("acc:",accuracy_score(ztest,zpredict))
```
