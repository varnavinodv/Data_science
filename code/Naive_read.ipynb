```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
x=pd.read_csv('iris.csv')
print(x.info())
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print(x.describe())
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print(x.shape)
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print(type(x))
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
print(x.head(5))
print(x.tail(3))
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
data=x.dropna()
print(data)
print(x.duplicated().any())
print(x.drop_duplicates(inplace=True))
print(x)
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
y=x.iloc[:,0:4]
z=x.iloc[:,4]
le=LabelEncoder()
zn=le.fit_transform(z)
print(zn)
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
ytrain,ytest,ztrain,ztest=train_test_split(y,zn,test_size=.25,random_state=25)
print(ztest)
nb=GaussianNB()
nb.fit(ytrain,ztrain)
zpredict=nb.predict(ytest)
print("test cls" ,ztest)
print("prdctcls:",zpredict)
print("acc:",accuracy_score(ztest,zpredict,normalize=True))
# print(nb.predict_proba(ytest)[:5])
cm=confusion_matrix(ztest,zpredict)
print(cm)
cm_dis=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=le.classes_)
cm_dis.plot()
plt.show()
```
