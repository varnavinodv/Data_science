```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
x=pd.read_csv('iris.csv')
print(x.info())
print(x.describe())
print(x.head(3))
print(x.tail(3))
print(type(x))
data=x.dropna()
print(data)
print(x.duplicated().any())
print(x.drop_duplicates())
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
y=x.iloc[:,:4].values
z=x.iloc[:,4].values
print("data",y)
print("targt",z)
le=LabelEncoder()
zn=le.fit_transform(z)
print(zn)
ytrain,ytest,ztrain,ztest=train_test_split(y,zn,test_size=.25,random_state=25)
print(ztest)
ln=LinearRegression()
ln.fit(ytrain,ztrain)
zpredict=ln.predict(ytest)
print("testcls:-----",ztest)
print("prdcls:------",zpredict)
print("r2:",r2_score(ztest,zpredict))
print("MSerr:",mean_squared_error(ztest,zpredict))
print("coef:",ln.coef_)
print("intscpt:",ln.intercept_)

```
