```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
iris=datasets.load_iris()
y=iris.data
z=iris.target
# print(y)
# print(z)
print("ftrs:",iris.feature_names)
print("trgt:",iris.target_names)
print(iris.DESCR)
ytrain,ytest,ztrain,ztest=train_test_split(y,z,test_size=.25,random_state=1)
print(ztest)
dt=DecisionTreeClassifier(criterion='entropy',splitter='best',max_depth=None,min_samples_split=2,min_samples_leaf=1)
dt.fit(ytrain,ztrain)
zpredict=dt.predict(ytest)
print("testcls:-----",ztest)
print("prdcls:------",zpredict)
print("acc:",accuracy_score(ztest,zpredict))
cm=confusion_matrix(ztest,zpredict)
print(cm)
cm_dis=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=iris.target_names)
cm_dis.plot()
plt.show()
dc=plt.figure(figsize=(10,10))
tree.plot_tree(dt)
plt.show(dc)

```
