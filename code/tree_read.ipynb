```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
x=pd.read_csv('iris.csv')
print(x.info())
print(x.describe())
print(x.head(3))
print(x.tail(3))
print(type(x))
data=x.dropna()
print(data)
print(x.duplicated().any())
print(x.drop_duplicates())
print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
y=x.iloc[:,:4].values
z=x.iloc[:,4].values
print("data",y)
print("targt",z)
le=LabelEncoder()
zn=le.fit_transform(z)
print(zn)
ytrain,ytest,ztrain,ztest=train_test_split(y,zn,test_size=.25,random_state=1)
print(ztest)
dt=DecisionTreeClassifier(criterion='entropy',splitter='best',max_depth=None,min_samples_split=2,min_samples_leaf=1)
dt.fit(ytrain,ztrain)
zpredict=dt.predict(ytest)
print("testcls:-----",ztest)
print("prdcls:------",zpredict)
print("acc:",accuracy_score(ztest,zpredict))
cm=confusion_matrix(ztest,zpredict)
print(cm)
cm_dis=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=le.classes_)
cm_dis.plot()
plt.show()
dc=plt.figure(figsize=(10,10))
tree.plot_tree(dt)
plt.show(dc)

```
