```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from PIL import Image
a=Image.open('dog.jpg') 
plt.imshow(a)
b=np.array(a)
c=ndimage.rotate(b,90)
plt.imshow(c)





```

```python
z=ndimage.zoom(b,(0.15,0.15,1))
plt.imshow(z)
```

```python
#print(b.shape)
#h,w,c=b.shape
#print(h,w,c)
p=np.array([[1,0.5,0],[0,1,0],[0,0,1]])
x=ndimage.affine_transform(b,p)
y=Image.fromarray(x)
plt.imshow(y)
```
