```python
import numpy as np
from numpy import random as r
import matplotlib.pyplot as plt
x=['varna','rilu','sani','vysh','sajin','amegh','abnv','neha','amrtsh','pooja']
y=r.randint(0,100,10)
plt.scatter(x,y,color='blue',marker='*',edgecolor='black',s=100)
y=r.randint(0,100,10)
plt.scatter(x,y,color='red',edgecolor='black')
plt.xlabel("X-Axis")
plt.ylabel("y-Axis")
plt.title('marks',color="green")
plt.legend(labels = ('physics', 'chemistry')) 
```
