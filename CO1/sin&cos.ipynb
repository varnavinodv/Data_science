```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, 5*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y,color="brown")
z=np.cos(x)
plt.plot(x,z,color="green")
plt.xlabel("X-Axis")
plt.ylabel("y-Axis")
plt.title('WAVES',color="red")
plt.legend(labels = ('sin', 'cos')) 
plt.show()
```
