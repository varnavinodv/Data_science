```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, 5*np.pi, 0.1)
y = np.sin(x)
plt.subplot(1,3,1)
plt.plot(x,y,color="brown")
z=np.cos(x)
plt.subplot(1,3,2)
plt.plot(x,z,color="green")
a=np.tan(x)
plt.subplot(1,3,3)
plt.plot(x,a,color="yellow")
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(x,y,color="brown")
plt.subplot(3,1,2)
plt.plot(x,z,color="green")
plt.subplot(3,1,3)
plt.plot(x,a,color="yellow")
plt.show()
```
